```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
drugData = pd.read_csv(mouse_drug_data_to_load, encoding = "ISO-8859-1")
clinicalData = pd.read_csv(clinical_trial_data_to_load, encoding = "ISO-8859-1")
# Combine the data into a single dataset
tumorData = pd.merge(clinicalData,drugData,how='inner',on='Mouse ID')


tumorData = tumorData [(tumorData["Drug"] == 'Capomulin') |
                       (tumorData["Drug"] == 'Infubinol') |
                       (tumorData["Drug"] == 'Ketapril') |
                       (tumorData["Drug"] == 'Placebo')
                      ]
# Display the data table for preview
tumorData.head()
tumorData.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
# Convert to DataFrame
meanTumor = pd.DataFrame(tumorData.groupby(['Drug','Timepoint']).mean()['Tumor Volume (mm3)'])
#meanTumor.reset_index()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
# Convert to DataFrame
# Preview DataFrame

stdErrTumor = pd.DataFrame(tumorData.groupby(['Drug','Timepoint']).sem()['Tumor Volume (mm3)'])
stdErrTumor.reset_index()
```

```python
#meanTumor is stacked data (hierarchial index).
# Each row corresponds to one row. 
#With DataFrame.stack and DataFrame.unstack, we can toggle between
#hierarchical indices and hierarchical columns. 
#unstack to hierarchial columns
# see https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.unstack.html

#meanTumor is now hierarchial columns

# MultiIndex(levels=[['Tumor Volume (mm3)'], ['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']],
#            labels=[[0, 0, 0, 0], [0, 1, 2, 3]],
#            names=[None, 'Drug'])
meanTumor = meanTumor.unstack(level = 0)
meanTumor.columns = meanTumor.columns.get_level_values(1)
meanTumor

#####   **************** IMPORTANT  ***************  ############
#need to do this to match or align with meanTumor in plot
stdErrTumor = stdErrTumor.unstack(level = 0)
stdErrTumor.columns = stdErrTumor.columns.get_level_values(1)
stdErrTumor
##############  dont delete above  #############################
```

```python
meanTumor.columns  #this assignment focuses only 4 treatments
```

```python
xvals = meanTumor.index #timepoints

#nice information on markers
# https://stackoverflow.com/questions/8409095/matplotlib-set-markers-for-individual-points-on-a-line
markers = ['o','^','s','d']

#information on colors
# https://images.search.yahoo.com/search/images;_ylt=AwrEwhUy1qpcWs8AROBXNyoA;_ylu=X3oDMTEycWNsbGxlBGNvbG8DYmYxBHBvcwMxBHZ0aWQDQjY4OThfMQRzZWMDc2M-?p=color+matplotlib+pandas&fr=mcafee#id=2&iurl=https%3A%2F%2Fpython-graph-gallery.com%2Fwp-content%2Fuploads%2F100_Color_names_python.png&action=click
colors = ['r','b','g','grey']

pos = 0

fig, axis = plt.subplots()

for col in meanTumor.columns:   
    axis.errorbar(xvals,meanTumor[col],stdErrTumor[col],
                  marker = markers[pos],
                  color = colors[pos],
                  linestyle = '--', 
                  capthick = 1,
                  capsize = 2
                 )
 
    pos += 1  #tracker for colors and markers

axis.set_xlim(0, xvals.max()+1)
axis.set_xlabel("Time (Days)")

axis.set_ylim(min(meanTumor.min()-3), max(meanTumor.max()+3))
axis.set_ylabel("Tumor Volumes (mm3)")

axis.grid()
plt.legend(loc="best")
plt.title('Tumor Response to Treatment')
plt.figure(figsize = (100,50))

plt.show()
```

```python
plt.savefig("./Images/tumorVolumeResponse.png")
```

## Metastatic Response to Treatment

```python
# Minor Data Munging to Re-Format the Data Frames
meanMetastatic = pd.DataFrame(tumorData.groupby(['Drug','Timepoint']).mean()['Metastatic Sites'])
# Preview that Reformatting worked
meanMetastatic.reset_index()

```

```python
stdErrMetastatic = tumorData.groupby(['Drug','Timepoint']).sem()['Metastatic Sites']
stdErrMetastatic.reset_index()
stdErrMetastatic.head()
```

```python
meanMetastatic = meanMetastatic.unstack(level = 0)
meanMetastatic.columns = meanMetastatic.columns.get_level_values(1)
# meanMetastatic.columns
```

```python
#####   **************** IMPORTANT  ***************  ############
#need to do this to match or align with meanMetastatic in plot
stdErrMetastatic = stdErrMetastatic.unstack(level = 0)
stdErrMetastatic.columns = stdErrMetastatic.columns.get_level_values(0)
stdErrMetastatic.head()
```

```python
# Generate the Plot (with Error Bars)
# Save the Figure
# Show the Figure
```

```python

fig, axisM = plt.subplots()

pos = 0
for col in meanMetastatic.columns:   
    axisM.errorbar(xvals,meanMetastatic[col],stdErrMetastatic[col],
                  marker = markers[pos],
                  color = colors[pos],
                  linestyle = '--', 
                  capthick = 1,
                  capsize = 2
                 )
    # print(col,pos)  #debugging and testing
    pos += 1  #tracker for colors and markers

axisM.set_xlim(0, xvals.max()+1)
axisM.set_xlabel("Time (Days)")

axisM.set_ylim(min(meanMetastatic.min()), max(meanMetastatic.max()))
axisM.set_ylabel("Metastatic Sites")

axisM.grid()
plt.legend(loc="best")
plt.title('Metastatic Sites')
plt.figure(figsize = (100,50))

plt.show()
```

## Summary Bar Graph

```python
meanTumor.head()
```

```python
# Calculate the percent changes for each drug
# Display the data to confirm
```

```python
#calculates % tumor volume change for each drug
lastTreatment = meanTumor.loc[45, :]
begTreatment = meanTumor.loc[0,:]
tumorPctChange = ((lastTreatment - begTreatment)/begTreatment) * 100
tumorPctChange.index
```

```python
# Store all Relevant Percent Changes into a Tuple
# Splice the data between passing and failing drugs
# Orient widths. Add labels, tick marks, etc. 
# Use functions to label the percentages of changes
# Call functions to implement the function calls
# Save the Figure
fig, ax = plt.subplots()
x_axis = tumorPctChange.index
y_axis = tumorPctChange
plt.title('Tumor Volume Change over 45 Day Treatment')
plt.ylabel('% Tumor Value Change')
plt.ylim(-40, 60)


plt.bar(np.arange(4),tumorPctChange,
        color = ['red' if tumorPctChange[col] > 0 else 'green' for col in np.arange(len(x_axis))])

plt.xticks(np.arange(len(x_axis)), x_axis)

#count for x coordinate location of value label
count = 0

#loops through tumor change data and places the value in the correct position based on + or - values
for r in tumorPctChange:
    if r < 0:
        y_coor = -4.5
    else:
        y_coor = 2
    plt.text(count, y_coor, str(round(r, 1)) + '%', multialignment = 'center', color = 'white')
    count += 1
    
# Show the Figure

plt.grid()
plt.show()
```
